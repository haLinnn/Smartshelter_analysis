```python
import pandas as pd
import numpy as np
from haversine import haversine, Unit
import openpyxl
```

```python
subway_df = pd.read_csv('../data/전처리데이터/부평구 지하철.csv',encoding='cp949')
subway_df
```

```python
bus_df= pd.read_csv('../data/전처리데이터/부평구 승차인원 추가.csv',encoding='cp949')
bus_df
```

```python
bus_df.drop(['Unnamed: 0'], axis = 1, inplace = True)
bus_df
```

```python
place_df = pd.read_csv('../data/부평구_버스정류장_위치.csv',encoding='cp949')
place_df
```

```python
place_df.rename(columns = {'정류장_1':'정류소번호'}, inplace=True)
place_df
```

```python
merge = pd.merge(bus_df, place_df, on = '정류소번호', how = 'inner')
merge
```

```python
merge.drop(['정류소','정류장_2'], axis = 1, inplace = True)
merge
```

```python
subway_df
```

```python
distance_compare_min = []
min_distance_staion = []

## for문으로 비교
for i in range(len(merge)):
    bus_stop = (merge['위도'][i], merge['경도'][i])
    distance = 999999
    near_station = ''
    for j in range(len(subway_df)):
        station = (subway_df['sub_lat'][j], subway_df['sub_lng'][j])
        tmp_distance = np.round(float(haversine(bus_stop, station, unit='m')), 3)
        if tmp_distance < distance:
            distance = tmp_distance
            near_station = subway_df['역명'][j]
    distance_compare_min.append(distance)
    min_distance_staion.append(near_station)
    # print(bus_df['정류소명'][i], near_station, distance)
```

```python
merge['가장 가까운 지하철역'] = min_distance_staion
merge['지하철역 최단거리(m)'] = distance_compare_min
```

```python
merge.head()
```

```python
merge.to_csv("../data/전처리데이터/부평구 최종 데이터.csv", encoding='cp949')
```
