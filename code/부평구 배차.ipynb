**버스 배차간격 데이터 불러오기**

```python
term = pd.read_csv('../data/전처리데이터/부평구 버스 배차간격_수정.csv',encoding='cp949')
term
```

```python
term.drop(['경유지'], axis=1, inplace=True)
term
```

```python
# 딕셔너리 만들기
dic = dict(term.values.tolist())
dic
```

```python
s=0
cnt=0
bus_avg = []
a = bus_df['노선번호2']
for i in a:
    s=0
    cnt=0
    for j in i: # j는 버스 번호
        if j in dic.keys():
            s += dic[j]
            cnt=cnt+1
    bus_avg.append(format(s/cnt, ".2f")) # 소수 둘째자리까지 표현
    print(s)
```

```python
bus_avg
```

```python
bus_df['평균 배차 간격'] = bus_avg
bus_df
```

```python
# 최대 배차 구하기
max=0
bus_max = []
a = bus_df['노선번호2']
for i in a:
    max=0
    for j in i: # j는 버스 번호
        if j in dic.keys():
            if dic[j] > max:
                max = dic[j]
    bus_max.append(max)
    print(max)
```

```python
bus_df['최대 배차 간격'] = bus_max
bus_df
```

```python
bus_df.drop(['노선번호'], axis=1, inplace=True)
bus_df
```

```python
bus_df.rename(columns = {'노선번호2':'노선번호'}, inplace=True)
bus_df
```

```python
bus_df.to_csv('../data/전처리데이터/부평구 정류장명 및 위치와 버스.csv', encoding = 'cp949', index=None)
```
