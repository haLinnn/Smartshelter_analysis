**버스 배차 간격 데이터 불러오기**

```python
term = pd.read_csv('../data/전처리데이터/성동구 버스 배차간격.csv', encoding='cp949')
term
```

```python
df.style.hide_index()
```

```python
# 딕셔너리 만들기
dic = dict(term.values.tolist())
dic
```

```python
for i in df['버스 노선 리스트2']:
    print(i)
```

```python
for i in df['버스 노선 리스트2']:
    for j in i: # j는 버스 번호
        print(j)
```

```python
s=0
cnt=0
bus_avg = []
a = df['버스 노선 리스트2']
for i in a:
    s=0
    cnt=0
    for j in i: # j는 버스 번호
        if j in dic.keys():
            s += dic[j]
            cnt=cnt+1
    bus_avg.append(format(s/cnt, ".2f")) # 소수 둘째자리까지 표현
    print(s)
```

```python
# dic['9403']
```

```python
bus_avg
```

```python
df['평균 배차 간격'] = bus_avg
df
```

```python
# 최대 배차 구하기
max=0
bus_max = []
a = df['버스 노선 리스트2']
for i in a:
    max=0
    for j in i: # j는 버스 번호
        if j in dic.keys():
            if dic[j] > max:
                max = dic[j]
    bus_max.append(max)
    print(max)
```

```python
df['최대 배차 간격'] = bus_max
df
```
