```python
import pandas as pd
import numpy as np
import pickle
```

```python
df = pd.read_csv('../data/2021년_버스노선별_정류장별_시간대별_승하차_인원_정보.csv', encoding='cp949')
# df = pd.read_csv('../data/2022년_버스노선별_정류장별_시간대별_승하차_인원_정보(06월).csv', encoding='cp949')
df.rename(columns={'?"사용년월"':'사용년월'}, inplace=True)
print(len(df))
df.iloc[:,:6].tail(10)
```

```python
with open('../data/전처리데이터/sungdong_busstop.pickle', 'rb') as fr:
    sungdong_busstop = pickle.load(fr)
```

```python
df = df[df['표준버스정류장ID'] // 1000000 == 103]  # 성동구에 위치하는 버스정류장 모음
df.reset_index(drop=True, inplace = True)
print(len(df))
df.head()
```

```python
processing_df = pd.DataFrame(columns = df.columns)
cnt = 0
for idx in df.index:
    busstop_id = df.iloc[idx]['표준버스정류장ID']
    if busstop_id in sungdong_busstop:
        processing_df.loc[cnt] = df.loc[idx]
        cnt += 1
```

```python
busstop_id = processing_df.iloc[0]['표준버스정류장ID']
total = {} # dict
arr = [] # 정류장별 지나는 버스번호 리스트
for idx in processing_df.index:
    if busstop_id != processing_df.iloc[idx]['표준버스정류장ID']:
        if busstop_id in list(total.keys()):
            total[busstop_id] += arr
        else:
            total[busstop_id] = arr
        arr = []
        busstop_id = processing_df.iloc[idx]['표준버스정류장ID']
    arr.append(processing_df.iloc[idx]['노선번호'])
total
```

```python
# 버스정류장 수
len(total)
```

```python
k = list(set(processing_df['표준버스정류장ID'].tolist()))
len(k)
```

```python
sungdong_bus = pd.DataFrame(columns = {'버스정류장', '버스 노선 리스트'})
cnt = 0
busstop = []
buslist = []
for k, v in total.items():
    busstop.append(k)
    buslist.append(list(set(v))) # 월별로 표기되어있으므로 중복값 제거
    cnt += 1
```

```python
a=[]
for i in range(len(buslist)):
    cnt=0
    for j in range(len(buslist[i])):
        cnt=cnt+1
    a.append(cnt)
a
```

```python
sungdong_bus['버스정류장'] = busstop
sungdong_bus['버스 노선 리스트'] = buslist
sungdong_bus['버스 수'] = a
```

```python
sungdong_bus
```

```python
sungdong_bus.to_csv('../data/전처리데이터/성동구 버스정류장별 버스 목록.csv', encoding='cp949')
```
