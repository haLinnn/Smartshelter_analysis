```python
import pandas as pd
import numpy as np
```

## 생활인구

```python
population_df = pd.read_csv('../data/성동구_버스정류장_인구.csv', encoding='cp949')
df = pd.read_csv('../data/전처리데이터/성동구 버스 진짜 최종.csv', encoding='cp949')
df = df[['도로명', '도로폭', '도로폭_숫자']]
df2 = pd.read_csv('../data/전처리데이터/성동구 쉼터여부_버스 승차수_버스수_배차간격_지하철역최단거리.csv', encoding='cp949')
df2.drop(['Unnamed: 0'], axis = 1, inplace = True)
population_df.head()
```

```python
sungdong_df = pd.concat([df2, df], axis=1)
sungdong_df = pd.concat([sungdong_df, population_df['인구']], axis=1)
sungdong_df.head()
```

```python
sungdong_df.to_csv('../data/전처리데이터/성동구 최종 데이터_수정.csv', encoding='cp949')
```
