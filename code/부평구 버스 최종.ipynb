### 앞 파일에 도로폭 추가

```python
import pandas as pd
import numpy as np

df = pd.read_csv('../data/인천광역시_차로폭, 차로수_현황_20211231.csv',encoding='cp949')
df
```

```python
df.drop(['번호','기점','종점','비고', '연장(미터)','면적(제곱미터)'], axis=1, inplace=True)
df
```

```python
df.info()
```

```python
df.head(17)
```

```python
df['폭원(미터)'] = pd.to_numeric(df['폭원(미터)'])
```

```python
df.rename(columns = {'폭원(미터)':'도로폭'}, inplace=True)
```

```python
df['도로폭_숫자'] = df['도로폭']
df
```

```python
df.loc[df['도로폭_숫자'] < 6, '도로폭_숫자'] = 1
df.loc[df['도로폭_숫자'].between(6,8,inclusive='left'),'도로폭_숫자'] = 2
df.loc[df['도로폭_숫자'].between(8,10,inclusive='left'),'도로폭_숫자'] = 3
df.loc[df['도로폭_숫자'].between(10,12,inclusive='left'),'도로폭_숫자'] = 4
df.loc[df['도로폭_숫자'].between(12,15,inclusive='left'),'도로폭_숫자'] = 5
df.loc[df['도로폭_숫자'].between(15,20,inclusive='left'),'도로폭_숫자'] = 6
df.loc[df['도로폭_숫자'].between(20,25,inclusive='left'),'도로폭_숫자'] = 7
df.loc[df['도로폭_숫자'].between(25,30,inclusive='left'),'도로폭_숫자'] = 8
df.loc[df['도로폭_숫자'].between(30,35,inclusive='left'),'도로폭_숫자'] = 9
df.loc[df['도로폭_숫자'].between(35,40,inclusive='left'),'도로폭_숫자'] = 10
df.loc[df['도로폭_숫자'].between(40,50,inclusive='left'),'도로폭_숫자'] = 11
df.loc[df['도로폭_숫자'].between(50,70,inclusive='left'),'도로폭_숫자'] = 12
df.loc[df['도로폭_숫자'] > 70, '도로폭_숫자'] = 13
df
```

```python
df.to_csv('../data/전처리데이터/부평구 도로폭.csv', encoding = 'cp949', index=None)
```

### 버스수

```python
df2 = pd.read_csv('../data/전처리데이터/부평구 정류장명 및 위치와 버스_최종.csv',encoding='cp949')
df2
```

```python
df2['노선번호'] = df2['노선번호'].str.split(', ').to_list()
```

```python
for i in df2['노선번호']:
    for j in i:
        print(j)
```

```python
a=[]
for i in range(len(df2['노선번호'])):
    cnt=0
    for j in range(len(df2['노선번호'][i])):
        cnt=cnt+1
    a.append(cnt)
a
```

```python
df2['버스수'] = a
```
